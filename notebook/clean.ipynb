## Start working on tables:

```python
#join df_1 & df_2 table into table_3
table_3 = pd.concat([df_1, df_2], axis=0)
display(table_3)
```

```python
# initial table analysis to get familiar
display(table_3.head(), f"Table shape: {table_3.shape}", table_3.dtypes)
```

```python
table_3['process_step'].unique()
```

```python
# handle the missing values in table_1
```

```python
# Fill NaN values in categorical columns with 'undefined'
table_1['gendr'] = table_1['gendr'].fillna('undefined')
table_1.isna().sum()
```

```python
table_1 = table_1.dropna(subset=['clnt_tenure_yr', 'clnt_tenure_mnth', 'clnt_age', 'num_accts', 'bal', 'calls_6_mnth', 'logons_6_mnth'])
display(table_1.isna().sum())
```

```python
# handle the missing values in table 2 => transform the NaN to Undefined
```

```python
table_2['Variation'].unique()
```

```python
# Replace NaN values with 'undefined'
table_2.fillna('undefined', inplace=True)
display(table_2.head(), table_2.isna().sum())
```

```python
# date and time parsing in table_3
table_3['date_time'] = pd.to_datetime(table_3['date_time'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
```

```python
# Extract year, date, and time from 'date_time' column
table_3['year'] = table_3['date_time'].dt.year
table_3['date'] = pd.to_datetime(table_3['date_time'].dt.date, format='%Y-%m-%d', errors='coerce')
table_3['time'] = pd.to_datetime(table_3['date_time'].dt.time.astype(str), format='%H:%M:%S', errors='coerce').dt.time
```

```python
# Display the updated table with the new columns
display(table_3.head())
```
