```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv("../data/clean/final_df.csv")
display(df, df.describe(), df.dtypes)
```

```python
# Ensure date_time is in datetime format
df['date_time'] = pd.to_datetime(df['date_time'])

# Confirm data types
print(df.dtypes)
```

## Time spent on each step: 

```python
# Define the age bins
age_bins = [30, 50, 70]
age_labels = ['30-50', '50-70']

# Create a new column for age group
df['age_group'] = pd.cut(df['clnt_age'], bins=age_bins, labels=age_labels, right=False)

# Filter the data for age groups 30-50 and 50-70
filtered_df = df[df['age_group'].isin(['30-50', '50-70'])]
display(filtered_df)
```

# Time Spent on Each Step by Group (Test vs Control)

```python
# Filter the data for age groups 30-50 and 50-70
df_filtered = df[(df['clnt_age'] >= 30) & (df['clnt_age'] <= 70)]
df_filtered['age_group'] = pd.cut(df_filtered['clnt_age'], bins=[30, 50, 70], labels=['30-50', '50-70'], right=False)

# Sort the data by client_id, visit_id, and date_time for proper time analysis
df_sorted = df_filtered.sort_values(by=['client_id', 'visit_id', 'date_time'])

# Calculate time differences for each client between steps
df_sorted['time_diff'] = df_sorted.groupby(['client_id', 'visit_id'])['date_time'].diff().dt.total_seconds()

# Drop rows where time_diff is NaN
df_sorted = df_sorted.dropna(subset=['time_diff'])

# Display the filtered data to verify the changes
df_sorted
```

---

## Statistical Tests:

Z-test/Chi-Square Test for Completion Rates

```python
from statsmodels.stats.proportion import proportions_ztest
import scipy.stats as stats
```

```python
# Example for 30-50 age group
test_30_50_completed = df_sorted[(df_sorted['Variation'] == 'Test') & (df_sorted['age_group'] == '30-50') & (df_sorted['process_step'] == 'confirm')].shape[0]
control_30_50_completed = df_sorted[(df_sorted['Variation'] == 'Control') & (df_sorted['age_group'] == '30-50') & (df_sorted['process_step'] == 'confirm')].shape[0]

# Total number of clients in the 30-50 age group
test_30_50_total = df_sorted[(df_sorted['Variation'] == 'Test') & (df_sorted['age_group'] == '30-50')].shape[0]
control_30_50_total = df_sorted[(df_sorted['Variation'] == 'Control') & (df_sorted['age_group'] == '30-50')].shape[0]

# Z-test for 30-50 group
count_30_50 = [test_30_50_completed, control_30_50_completed]
nobs_30_50 = [test_30_50_total, control_30_50_total]
z_stat_30_50, p_value_30_50 = proportions_ztest(count_30_50, nobs_30_50)

print(f"30-50 Age Group: Z-statistic: {z_stat_30_50}, P-value: {p_value_30_50}")
```

```python
if p_value_30_50 < 0.05:
    print("Reject the null hypothesis: There is a significant difference in completion rates.")
else:
    print("Fail to reject the null hypothesis: No significant difference in completion rates.")
```

```python
# Example for 50-70 age group
test_50_70_completed = df_sorted[(df_sorted['Variation'] == 'Test') & (df_sorted['age_group'] == '50-70') & (df_sorted['process_step'] == 'confirm')].shape[0]
control_50_70_completed = df_sorted[(df_sorted['Variation'] == 'Control') & (df_sorted['age_group'] == '50-70') & (df_sorted['process_step'] == 'confirm')].shape[0]

# Total number of clients in the 50-70 age group
test_50_70_total = df_sorted[(df_sorted['Variation'] == 'Test') & (df_sorted['age_group'] == '50-70')].shape[0]
control_50_70_total = df_sorted[(df_sorted['Variation'] == 'Control') & (df_sorted['age_group'] == '50-70')].shape[0]

# Z-test for 50-70 group
count_50_70 = [test_50_70_completed, control_50_70_completed]
nobs_50_70 = [test_50_70_total, control_50_70_total]
z_stat_50_70, p_value_50_70 = proportions_ztest(count_50_70, nobs_50_70)

print(f"50-70 Age Group: Z-statistic: {z_stat_50_70}, P-value: {p_value_50_70}")
```

```python
if p_value_50_70 < 0.05:
    print("Reject the null hypothesis: There is a significant difference in completion rates.")
else:
    print("Fail to reject the null hypothesis: No significant difference in completion rates.")
```

Result of the test:
- For the 30-50 age group, the new design (Test group) resulted in a higher completion rate.
- For the 50-70 age group, the completion rate did not significantly differ between the Test and Control groups.

5% test

```python
# Calculate completion rates
completion_rate_test_30_50 = test_30_50_completed / test_30_50_total
completion_rate_control_30_50 = control_30_50_completed / control_30_50_total
completion_rate_diff_30_50 = completion_rate_test_30_50 - completion_rate_control_30_50
```

```python
print(f"Test Group Completion Rate (30-50): {completion_rate_test_30_50:.4f}")
print(f"Control Group Completion Rate (30-50): {completion_rate_control_30_50:.4f}")
print(f"Completion Rate Difference (Test - Control): {completion_rate_diff_30_50:.4f}")
```

```python
# Check if completion rate increase is 5% or more
if completion_rate_diff_30_50 >= 0.05:
    print("The completion rate increase meets or exceeds the 5% threshold.")
else:
    print("The completion rate increase does not meet the 5% threshold.")
```

```python
completion_rate_test_50_70 = test_50_70_completed / test_50_70_total
completion_rate_control_50_70 = control_50_70_completed / control_50_70_total
completion_rate_diff_50_70 = completion_rate_test_50_70 - completion_rate_control_50_70
```

```python
print(f"Test Group Completion Rate (50-70): {completion_rate_test_50_70:.4f}")
print(f"Control Group Completion Rate (50-70): {completion_rate_control_50_70:.4f}")
print(f"Completion Rate Difference (Test - Control): {completion_rate_diff_50_70:.4f}")
```

```python
if completion_rate_diff_50_70 >= 0.05:
    print("The completion rate increase meets or exceeds the 5% threshold.")
else:
    print("The completion rate increase does not meet the 5% threshold.")
```

ANOVA for Time Spent on Each Step

```python
# Time spent for age group 30-50 and 50-70
grouped_time_30_50 = df_sorted[(df_sorted['age_group'] == '30-50')].groupby('Variation')['time_diff'].apply(list)
grouped_time_50_70 = df_sorted[(df_sorted['age_group'] == '50-70')].groupby('Variation')['time_diff'].apply(list)

# ANOVA for 30-50 age group
f_stat_30_50, p_value_30_50 = stats.f_oneway(grouped_time_30_50['Test'], grouped_time_30_50['Control'], grouped_time_30_50['undefined'])
print(f"30-50 Age Group: F-statistic: {f_stat_30_50}, P-value: {p_value_30_50}")

# ANOVA for 50-70 age group
f_stat_50_70, p_value_50_70 = stats.f_oneway(grouped_time_50_70['Test'], grouped_time_50_70['Control'], grouped_time_50_70['undefined'])
print(f"50-70 Age Group: F-statistic: {f_stat_50_70}, P-value: {p_value_50_70}")
```

```python
# Interpretation
if p_value_30_50 < 0.05:
    print("Reject the null hypothesis: There is a significant difference in time spent between the groups.")
else:
    print("Fail to reject the null hypothesis: No significant difference in time spent between the groups.")
```

```python
# Interpretation
if p_value_50_70 < 0.05:
    print("Reject the null hypothesis: There is a significant difference in time spent between the groups.")
else:
    print("Fail to reject the null hypothesis: No significant difference in time spent between the groups.")
```

Result of test:
- For the 30-50 age group, the time spent on each step differs significantly between the Test, Control, and Undefined groups.
- For the 50-70 age group, no significant difference was found in the time spent across the groups.
